```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_csv("./grafici-dataset.csv", sep=",", decimal=".")
data.head()
```

## Diagramma a barre

```python
data["genere"].value_counts().plot.bar(color=['pink', 'blue'])
plt.show()
```

## Diagramma a torta (aerogramma)

```python
data["genere"].value_counts().plot.pie(colors=['pink', 'blue'], legend=True, title="Genere partecipanti")
plt.show()
```

## Istogramma

```python
data["punteggio"].hist(bins=range(0, 43, 3))
plt.xticks(range(0, 43, 3))
plt.show()
```

```python
data["punteggio"].value_counts(bins=range(0, 43, 3), sort=False)
```

## Boxplot

```python
data["eta"].plot.box(vert=False)
```

## QQ-plot

```python
sample1 = data["punteggio"].sample(200)
sample2 = data["punteggio"].sample(200)

import statsmodels.api as sm

sm.qqplot_2samples(sample1, sample2, line='45')
plt.show()
```

## Scatterplot

```python
data.plot.scatter("tempo", "punteggio")
x = np.linspace(0, 40)
plt.plot(x, x, color="red", linestyle="dashed")
plt.xlim(-5, 50)
plt.ylim(-5, 50)
plt.show()
```

```python
data["punteggio"].corr(data["tempo"])
```
