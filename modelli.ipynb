## Bernoulliano

```python
from scipy.stats import bernoulli

p = 0.4
Be = bernoulli(p)
```

### Massa

```python
x = np.arange(2)

plt.vlines(x, 0, Be.pmf(x), linestyles="dashed")
plt.plot(x, Be.pmf(x), 'o')
plt.ylim(0, 1)
plt.grid()
plt.show()
```

### Ripartizione

```python
x = np.linspace(-0.1, 1.1, 1000)

plt.step(x, Be.cdf(x))
plt.ylim(-0.1, 1.1)
plt.grid()
plt.show()
```

## Binomiale

```python
from scipy.stats import binom

n = 5
p = 0.4
Bi = binom(n, p)
```

### Massa

```python
x = np.arange(10)

plt.vlines(x, 0, Bi.pmf(x), linestyles="dashed")
plt.plot(x, Bi.pmf(x), 'o')
plt.ylim(0, 1)
plt.grid()
plt.show()
```

### Ripartizione

```python
x = np.linspace(-1, 8, 1000)

plt.step(x, Bi.cdf(x))
plt.ylim(-0.1, 1.1)
plt.yticks(np.arange(0, 1.1, 0.1))
plt.xticks(np.arange(-1, 9))
plt.grid()
plt.show()
```

## Geometrico

```python
from scipy.stats import geom

p = 0.5
G = geom(p)
```

ATTENZIONE! La variabile aleatoria di modello geometrico usato da `scipy` assume il valore del primo successo (supporto $\lbrace 1, ..., +\infty \rbrace$), mentre a lezione contiamo solo il numero di insuccessi (supporto $\lbrace 0, ..., +\infty \rbrace$)

### Massa

```python
x = np.arange(0, 11)

plt.vlines(x, 0, G.pmf(x))
plt.plot(x, G.pmf(x), 'o')
plt.ylim(0, 1)
plt.yticks(np.arange(0, 1.1, 0.1))
plt.xticks(np.arange(0, 11))
plt.grid()
plt.show()
```

### Ripartizione

```python
x = np.linspace(0, 11, 1000)

plt.step(x, G.cdf(x))
plt.ylim(-0.1, 1.1)
plt.xticks(np.arange(0, 11))
plt.grid()
plt.show()
```

## Ipergeometrico

```python
from scipy.stats import hypergeom

estrazioni = 10
corretti = 20
sbagliati = 30

M = corretti + sbagliati # totale oggetti
n = corretti
N = estrazioni

H = hypergeom(M, n, N)
```

### Massa

```python
x = np.arange(0, 11)

plt.vlines(x, 0, H.pmf(x))
plt.plot(x, H.pmf(x), 'o')
plt.ylim(0, 1)
plt.yticks(np.arange(0, 1.1, 0.1))
plt.xticks(np.arange(0, 11))
plt.grid()
plt.show()
```

### Ripartizione

```python
x = np.arange(0, 10) # ATTENZIONE! non so perchè non funziona linspace o anche arange cambiando i parametri
# ATTENZIONE! a 0 la probabilità dovrebbe essere 0, non so come fixare

plt.step(x, H.cdf(x))
plt.ylim(-0.1, 1.1)
plt.grid()
plt.show()
```

## Uniforme continuo

```python
from scipy.stats import uniform

a = 1
b = 11
UnCo = uniform(a, b-a)
```

### Densità

```python
x = np.linspace(0, 12, 1000)

plt.plot(x, UnCo.pdf(x))
plt.ylim(0, 1)
plt.yticks(np.arange(0, 1.1, 0.1))
plt.xticks(np.arange(0, 13))
plt.grid()
plt.show()
```

### Ripartizione

```python
x = np.linspace(0, 12)

plt.plot(x, UnCo.cdf(x))
plt.ylim(-0.1, 1.1)
plt.xticks(range(0, 13))
plt.grid()
plt.show()
```

## Normale o Gaussiana

```python
from scipy.stats import norm

mu = 4
sigma = 12
N = norm(mu, sigma)
```

### Densità

```python
x = np.linspace(mu-50, mu+50, 1000)

plt.plot(x, N.pdf(x))
plt.axvline(mu, color="red", linestyle="dashed")
plt.grid()
plt.show()
```

### Ripartizione

```python
x = np.linspace(mu-50, mu+50, 1000)

plt.plot(x, N.cdf(x))
plt.axvline(mu, color="red", linestyle="dashed")
plt.ylim(-0.1, 1.1)
plt.yticks(np.arange(0, 1.1, 0.1))
plt.grid()
plt.show()
```
